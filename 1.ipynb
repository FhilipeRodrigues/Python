```python
# Aula 5 - Exercício A)
def salario_s(h,sem):
  rem_d = 18*h
  rem_s = rem_d*sem + 50
  return rem_s

salario_semanal = salario_s(4,5)
print('O estágiario ganha semanalmente: {} reais'.format(salario_semanal))
```
